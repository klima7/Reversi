```python
from copy import deepcopy
from time import time

import numpy as np

import board_tools as bt
from game_state import GameState
```

```python
def generate_all_states(size):
    all_states = []
    game_states = [GameState(size)]

    i = 0

    while game_states:
        game_state = game_states.pop(0)
        # if game_state.board in all_states:
        #     return
        all_states.append(game_state.board)
        legal_moves = game_state.get_moves()
        for legal_move in legal_moves:
            next_game_state = deepcopy(game_state)
            next_game_state.make_move(legal_move)
            game_states.append(next_game_state)

        i += 1
        if i % 200 == 0:
            not_zeros = np.sum(game_states[0].board != 0)
            # game_states[0].plot()
            print(i, not_zeros)


    return all_states
```

```python
all_states3 = generate_all_states(3)
print(all_states3[-1])
print(len(all_states3))
```

```python
start = time()
all_states = generate_all_states(4)
print(all_states[-1])
print(len(all_states))
print(time() - start)
```
